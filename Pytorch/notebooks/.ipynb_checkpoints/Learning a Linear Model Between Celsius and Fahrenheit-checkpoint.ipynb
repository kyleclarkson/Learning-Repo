# Learning a Linear Model Between Celsius and Fahrenheit

In this notebook be fit two parameters to a collection of temperater data. A scaler parameter $w$ and a basis parameter $b$. (i.e. we assume our (x,y) data matches an equation of the form $y= wx + b$ and we wish to determine these parameter values.)

We use gradient descent to determine these parameters by applying an L2 loss function to what our current model predicts and what our actual data is.


```python
%matplotlib inline

import torch as T
from matplotlib import pyplot as plt
```

```python
# Our data
# temperatures in unknown units
temp_x = [0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0]
# temperatures in celsius. 
temp_y = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
assert(len(temp_x) == len(temp_y))

# Create tensors for data
t_x = T.tensor(temp_x)
t_y = T.tensor(temp_y)
```

```python
# Plot data to see if our linear model is a "good idea"
fig = plt.figure(dpi=400)
plt.xlabel("Unknown Units")
plt.ylabel("Celsius")
plt.plot(t_x.numpy(), t_y.numpy(), 'o')
```

## Define model, loss function and gradient of loss function. 

Below define our model, loss (error) function, gradient of loss (error) function, and the partial derivatives of our loss function. Note the use of broadcasting by the tensor objects.  

Our model has two parameters, $w$ and $b$. The gradient of it is then given by the paritial derivatives with respective to both:  
$
\begin{align*}
    \nabla L &= \Big( \frac{\partial L}{\partial w}, \frac{\partial L}{\partial b}\Big)
\end{align*}
$

By applying the chain we can compute the partial derivatives with respect to each parameter. For example the partial derivative of the loss with respect to (wrt) $w$ is
$
\begin{align*}
   \frac{\partial L}{\partial w} &= \frac{\partial L}{\partial m}*\frac{\partial m}{\partial w}
\end{align*}
$  
where $m$ is the model. 

### Model (Linear)
$ y = w*x + b$
  
### Loss function
$ = \frac{1}{2}\sum_{i=1}^{n}(y_{pred_i} - y_{label_i})^2$  
$ = \frac{1}{2}\sum_{i=1}^{n}((w*x + b) - y_{label})^2$
  

### Gradient of Loss (wrt to model)
$
\begin{align*}
   \frac{1}{n} &= \sum_{i=1}^{n}(y_{pred} - y_{label})
\end{align*}
$  

### Gradient of  Model (wrt to w)
$
\begin{align*}
   \frac{\partial m}{\partial w} &= \frac{\partial y}{\partial w} = \frac{\partial}{\partial w} (wx + b)=w
\end{align*}
$
  
  
### Gradient of  Model (wrt to b)
$
\begin{align*}
   \frac{\partial m}{\partial b} &= \frac{\partial y}{\partial b} = \frac{\partial}{\partial b} (wx + b)=1
\end{align*}
$



```python
def model(x, w, b):
    return w*x + b

def loss_fn(y_pred, y_label):
    sq_error = (y_pred - y_label)**2
    return 0.5 * sq_error.mean()

def dloss_fn(y_pred, y_label):
    dsq = (y_pred - y_label) / y_pred.size(0)
    return dsq
    
def dmodel_dw(x, w, b):
    return x

def dmodel_db(x, w, b):
    return 1.0

def grad_fn(x, y_label, y_pred, w, b):
    dloss_dm = dloss_fn(y_pred, y_label)
    dloss_dw = dloss_dm * dmodel_dw(x, w, b)
    dloss_db = dloss_dm * dmodel_db(x, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])
```

```python
# Normalize data to be centered at 0 and scaled down by standard deviation. 
mean = t_x.mean()
std = t_x.std()
t_xnorm = t_x - mean
t_xnorm /= std
print("Mean: ", mean)
print("Standard Deviation: ", std)
print(t_x)
print(t_xnorm)
```

```python
# Training loop
def train(n_epochs, lr, params, x, y_label):
    for epoch in range(1, n_epochs+1):
        w, b = params
        
        y_pred = model(x, w, b)
        loss = loss_fn(y_pred, y_label)
        grad = grad_fn(x, y_label, y_pred, w, b)
        params = params - lr*grad
        
        if(epoch % 200 == 0):
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
            
    return params
```

```python
params = train(10000, 0.001, torch.tensor([1.0, 0]), t_xnorm, t_y)
```

```python
print(params)
```

```python
# Plot line resulting from the parameters learned. 
fig = plt.figure(dpi=400)
plt.xlabel("Unknown Units")
plt.ylabel("Celsius")
plt.plot(t_x.numpy(), model(t_xnorm, *params).detach().numpy())
plt.plot(t_x.numpy(), t_y.numpy(), 'o')
```
